```python
#일자 데이터 타입을 datetime으로 변경 
Changwon_Call_df['일자(YYYYMMDD)'] = Changwon_Call_df['일자(YYYYMMDD)'].astype('str')
Changwon_Call_df['일자(YYYYMMDD)'] = pd.to_datetime(Changwon_Call_df['일자(YYYYMMDD)'])
Changwon_Call_df.dtypes
```

```python
# 연령 데이터 타입을 object로 변경 
Changwon_Call_df['연령'] = Changwon_Call_df['연령'].astype('str')
Changwon_Call_df.dtypes
```

```python
#파생변수 요일 생성
def create_derived_variable_weekday(x,y):
    Changwon_Call_df[x] = 0 
    Changwon_Call_df[x] = Changwon_Call_df[y].map(lambda x : x.weekday())
    weekdays = ['월요일','화요일','수요일','목요일','금요일','토요일','일요일']
    Changwon_Call_df[x] = Changwon_Call_df[x].map(lambda x : weekdays[x])
    return Changwon_Call_df
```

```python
create_derived_variable_weekday('요일','일자(YYYYMMDD)')
```

```python
# 주중,주말,휴일 파생변수 생성하기 
def create_derived_variable_weekend(x,y,z,k):
    Changwon_Call_df[x] = 0
    Changwon_Call_df[y] = 0
    Changwon_Call_df[x] = Changwon_Call_df[z].map(lambda x : 1 if (x=='토요일' or x=='일요일') else 0)
    holidays = ['2019-01-01','2019-02-04','2019-02-05','2019-02-06','2019-03-01','2019-05-06','2019-05-12','2019-06-06','2019-08-15','2019-09-12','2019-09-13','2019-09-14','2019-10-03','2019-10-09','2019-12-25',
               '2020-01-01','2020-01-24','2020-01-25','2020-01-26','2020-01-27','2020-03-01','2020-04-30','2020-05-05','2020-06-06','2020-08-17']
    holidays = pd.to_datetime(holidays)
    Changwon_Call_df[y] = Changwon_Call_df[k].map(lambda x : 1 if (x in holidays) else 0)
    return Changwon_Call_df
```

```python
create_derived_variable_weekend('주말/주중','휴일','요일','일자(YYYYMMDD)')
# 주말 1 주중 0 
# 휴일 또는 대체 휴일 (빨간 날) 1 아니면 0 
```
