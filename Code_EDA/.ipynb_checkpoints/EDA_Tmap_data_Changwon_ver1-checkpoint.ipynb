```python
import requests
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import matplotlib.font_manager as fm 
fontpath='C:/Windows/Fonts/malgunbd.ttf'
font=fm.FontProperties(fname=fontpath,size=9)
```

```python
# 2019년 여행/레저의 검색지명을 리스트로 생성
mask = (Changwon_Tmap_df2['년도']==2019)&(Changwon_Tmap_df2['검색지유형1']=='여행/레저')
words_2019 = Changwon_Tmap_df2.loc[mask,:]['검색지명'].to_list()
len(words_2019)
```

```python
# 2020년 여행/레저의 검색지명을 리스트로 생성
mask = (Changwon_Tmap_df2['년도']==2020)&(Changwon_Tmap_df2['검색지유형1']=='여행/레저')
words_2020 = Changwon_Tmap_df2.loc[mask,:]['검색지명'].to_list()
len(words_2020)
```

```python
# 2019년 레저/여행 카테고리의 검색지명의 빈도수를 바탕으로 워드클라우드 생성
wordcloud=WordCloud(font_path=fontpath).generate(' '.join(words_2019))

fig=plt.figure(figsize=(16,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig('peng_wordcloud.png')
```

```python
# 2020년 레저/여행 카테고리의 검색지명의 빈도수를 바탕으로 워드클라우드 생성
wordcloud=WordCloud(font_path=fontpath).generate(' '.join(words_2020))

fig=plt.figure(figsize=(16,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig('peng_wordcloud.png')
```
